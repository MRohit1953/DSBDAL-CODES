```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from seaborn import load_dataset

# Titanic dataset
import seaborn as sns

# Load Titanic dataset from seaborn
data = sns.load_dataset("titanic")


# Tips dataset
tips = sns.load_dataset("tips")

```

```python
print(data.columns)

# Ensure the column name is correct
sns.countplot(data=data, x='survived')
plt.show()
```

```python
sex_counts = data['sex'].value_counts()
sex_counts.plot(kind="pie", autopct="%.2f")
plt.show()
```

```python
plt.hist(data['age'], bins=5)
plt.show()
```

```python
sns.distplot(data['age'])
plt.show()
```

```python
sns.scatterplot(x=tips["total_bill"], y=tips["tip"])
plt.show()
```

```python
sns.scatterplot(x=tips["total_bill"], y=tips["tip"], hue=tips["sex"])
plt.show()
```

```python
sns.barplot(x=data['pclass'], y=data['age'])
plt.show()
```

```python
sns.barplot(x=data['class'], y=data['fare'], hue=data["sex"])
plt.show()
```

```python
sns.boxplot(x=data['sex'], y=data['age'])
plt.show()
```

```python
sns.boxplot(x=data['sex'], y=data['age'], hue=data['survived'])
plt.show()
```

```python
sns.distplot(data[data['survived'] == 0]['age'], hist=False, color="blue")
sns.distplot(data[data['survived'] == 1]['age'], hist=False, color="orange")
plt.show()
```

```python
cross_tab = pd.crosstab(data['class'], data['survived'])
print(cross_tab)
```

```python
cross_tab = pd.crosstab(data['class'], data['survived'])
sns.heatmap(cross_tab)
plt.show()
```

```python
cross_tab = pd.crosstab(data['parch'], data['survived'])
sns.clustermap(cross_tab)
plt.show()
```
